# Support Vector Regression

```python
#Import Libraries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

dataset1 = pd.read_csv('sample/Position_Salaries.csv')
x1 = dataset1.iloc[:,1:-1].values
y1 = dataset1.iloc[:,-1].values

dataset1
```

```python
x1
```

```python
y1
```

```python
#Reshape the y value from rows to columns

y1 = y1.reshape(len(y1),1)

y1
```

#### Feature Scaling using standardscalar

Standardization scales each input variable separately by subtracting the mean (called centering) and dividing by the standard deviation to shift the distribution to have a mean of zero and a standard deviation of one.

Standardization can give values that are both positive and negative centered around zero.

It may be desirable to normalize data after it has been standardized.

```python
from sklearn.preprocessing import StandardScaler as sc

sc_x1 = sc()
sc_y1 = sc()
```

```python
x1 = sc_x1.fit_transform(x1)
y1= sc_y1.fit_transform(y1)
```

```python
x1
```

```python
y1
```

```python
#entire dataset will be trained for svr

from sklearn.svm import SVR
regressor_svr = SVR(kernel = 'rbf')
regressor_svr.fit(x1,y1)
```

```python
#Predict a new result

sc_y1.inverse_transform(regressor_svr.predict(sc_x1.transform([[6.5]])))
```

```python
#plot the svr

plt.scatter(sc_x1.inverse_transform(x1), sc_y1.inverse_transform(y1), color = 'red')
plt.plot(sc_x1.inverse_transform(x1), sc_y1.inverse_transform(regressor_svr.predict(x1)), color = 'blue')
plt.xlabel('Position level')
plt.ylabel('Salary')
plt.show()
plt.show()
```

# Decision Tree Regression

```python
# We are working with the same dataset as we used in svr

x2 = dataset1.iloc[:,1:-1].values
y2 = dataset1.iloc[:,-1].values
```

```python
x2
```

```python
y2
```

```python
from sklearn.tree import DecisionTreeRegressor as dtr
regressor_dtr = dtr()
regressor_dtr.fit(x2,y2)
```

```python
regressor_dtr.predict([[6.5]])
```

```python
#visualise the dtr
x_grid = np.arange(min(x2), max(x2), 0.01) #We need to plot at 0.01 intervals

x_grid
```

```python
x_grid = x_grid.reshape((len(x_grid), 1)) #reshaping it to one column
x_grid
```

```python
plt.scatter(x2,y2, color = 'red')
plt.plot(x_grid, regressor_dtr.predict(x_grid), color = 'blue') #insted of plotting for features, we will plot for x_grid
plt.xlabel('Position level')
plt.ylabel('Salary')
plt.show()
plt.show()
```

```python
# Random Forest and we are considering the same dataset as position salaries

x3 = dataset1.iloc[:,1:-1].values
y3 = dataset1.iloc[:,-1].values
```

```python
x3
```

```python
y3
```

```python
# Training random forest regressor for the whole model

from sklearn.ensemble import RandomForestRegressor as rfr
regressor_rfr = rfr(n_estimators = 10, random_state = 0) #n_estimators will be no of decision tree fed to model
regressor_rfr.fit(x3,y3)
```

```python
regressor_rfr.predict([[7.5]])
```

```python
# Visualising random forest

x_rfrgrid = np.arange(min(x3), max(x3), 0.1)
x_rfrgrid = x_rfrgrid.reshape((len(x_rfrgrid),1))

x_rfrgrid
```

```python
plt.scatter(x3,y3, color = 'red')
plt.plot(x_rfrgrid, regressor_rfr.predict(x_rfrgrid), color = 'purple')
plt.xlabel('Position level')
plt.ylabel('Salary')
plt.show()
plt.show()
```
