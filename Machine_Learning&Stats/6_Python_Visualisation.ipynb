```python
#Line chart
# A simple X,Y chart with no complex visualisation. 

import matplotlib.pyplot as plt #matplotlib has pyplot module and plt is an alias or reference

%matplotlib inline 
#this here is used only in jupyter notebook to display the graph inside the cell.

x = [1,2,3,4]
y = [10,20,30,40]

#line plot using function based. I have called pyplot function and specified graph as plot based.
plt.plot(x,y)

plt.show() 
#In Jupyter notebook, plt,show() will be default, but other IDE like spyder, google colab etc we need to specify plt.show()
```

```python
# say we want to plot two plot in single cells, we use subplots and also enter labels

plt.subplot(1,2,1) #rows, columns and plot number
plt.plot(x,y,'r') # r is a red color
plt.title('Sub Plot_1')
plt.xlabel('x')
plt.ylabel('y')

plt.subplot(1,2,2)
plt.plot(y,x,'g') # g is a red color
plt.title('Sub Plot_2')
plt.ylabel('y')
plt.xlabel('x')

plt.show()
# A simple line is not an effective, consider another example
```

```python
#3D line plot
from mpl_toolkits import mplot3d
import numpy as np
import matplotlib.pyplot as plt


fig = plt.figure()
ax = plt.axes(projection = '3d')

#line 3-D plot
Z = np.linspace(0,1,100)
x = Z * np.sin(30 * Z) #sine function on the z values assigned to x vertices
y = Z * np.cos(30 * Z) #cosine function on the z values assigned to y vertices

ax.plot3D(x,y,Z,'gray')
ax.set_title('3D line plot')
plt.show()
```

```python
#3D scatter plot

from mpl_toolkits import mplot3d #3D plot always this should come first
import numpy as np
import matplotlib.pyplot as plt
fig = plt.figure()
ax = plt.axes(projection='3d')
z = np.linspace(0, 1, 100)
x = z * np.sin(30 * z)
y = z * np.cos(30 * z)
c = x + y #c can be single color format string or sequence.
ax.scatter(x, y, z, c=c)
ax.set_title('3d Scatter plot')
plt.show()
```

# Seaborn - a library built on matplotlib. Used extensively for statistical plotting.

Seaborn has features.

    * Visualizing univariate and bivariate data
    * Fitting in and visualizing linear regression models
    * Plotting statistical time series data

types of graphs
 
    1. Relational plots: This plot is used to understand the relation between two variables. 
    2. Categorical plots: This plot deals with categorical variables and how they can be visualized.
    3. Distribution plots: This plot is used for examining univariate and bivariate distributions.
    4. Regression plots: The regression plots in seaborn are primarily intended to add a visual guide that helps to                    emphasize patterns in a dataset during exploratory data analyses.
    5. Multiplots - useful to draw conclusion on subset of dataset
    
http://seaborn.pydata.org/examples/index.html - Seaborn has more than 16 variants. We will cover top 5 or most used. Reader are requested atleast check the shared link once to see the magnitude of seaborn.

```python
# Distribution plot - Let us perform seabron viz on tips dataset

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

dataset = pd.read_csv('tips.csv')

dataset.head()
```

```python
#distplot 
sns.set_style('whitegrid')
sns.distplot(dataset['total_bill'], kde = False, color = 'red', bins = 30)
plt.show()

# KDE stands for Kernel Density Estimation and that is another kind of the plot in seaborn.
# bins is used to set the number of bins you want in your plot and it actually depends on your dataset.
# color is used to specify the color of the plot
```

```python
#jointplot - jointplot(x, y[, data, kind, stat_func, ...])   
```

```python
sns.jointplot(x = dataset['total_bill'], y = dataset['tip'], data = dataset)
plt.show()
```

```python
#using KDE

sns.jointplot(x = dataset['total_bill'], y = dataset['tip'], data = dataset, kind = 'kde')
plt.show()
```

```python
# pairplot - pairplot(data[, hue, hue_order, palette, ])    

sns.pairplot(dataset, hue = 'sex', palette ='rocket' ) # sex - male and female are given colors to effectively plot
plt.show()
```

```python
# rugplot - It plots datapoints in an array as sticks on an axis.Just like a distplot it takes a single column. 
#  Instead of drawing a histogram it creates dashes all across the plot.

#rugplot(a[, height, axis, ax]) 

sns.rugplot(dataset['total_bill'])
```

### Categorical Plots - Plots are basically used for visualizing the relationship between variables. 

    * Those variables can be either be completely numerical or a category like a group, class or division.

```python
# bar plot - aggregate the categorical data according to some methods and by defaults it is mean. 
# always choose categorical value for the x axis and numerical value for the y axis.
# barplot([x, y, hue, data, order, hue_order, ])

sns.barplot(x = 'smoker', y = 'total_bill', data = dataset, palette = 'plasma') 
# by default, the estimator which has numerical value has been performed statistical mean function on them.
```

```python
sns.barplot(x = 'smoker', y = 'total_bill', data = dataset, estimator = np.std) 
#here the estimator perform standard deviation using numpy libraries.
plt.show()
```

```python
# countplot - A countplot basically counts the categories and returns a count of their occurrences. 
# It is one of the most simple plots provided by the seaborn library.

#countplot([x, y, hue, data, order, ])

sns.countplot(x = 'sex', data = dataset)
```

```python
#box plots - A boxplot is sometimes known as the box and whisker plot.It shows the distribution of the quantitative data that represents the comparisons between variables. boxplot shows the quartiles of the dataset while the whiskers extend to show the rest of the distribution i.e. 
#the dots indicating the presence of outliers.

#boxplot([x, y, hue, data, order, hue_order, ])

sns.boxplot(x = 'day', y = 'total_bill', data = dataset, hue = 'sex')
plt.show()
```

#### from above box plot, we can infer that female spent more on saturday and male spent more on friday

```python
# violin plot - It is similar to the boxplot except that it provides a higher, more advanced visualization 
#   and uses the kernel density estimation to give a better description about the data distribution.
# violinplot([x, y, hue, data, order, ])

sns.violinplot(x = 'day', y = 'total_bill', data = dataset, hue = 'sex', split = True)
# split = True will give a broader understanding on the amount spent at diner based n gender
```

```python
sns.stripplot(x ='day', y ='total_bill', data = dataset,jitter = True, hue ='smoker', dodge = True) 
plt.show()
```

```python
# swarmplot - It is very similar to the stripplot except the fact that the points are adjusted so that they do not overlap.
#Some people also like combining the idea of a violin plot and a stripplot to form this plot. 
#One drawback to using swarmplot is that sometimes they dont scale well to really large numbers and takes a lot of computation to arrange them. 

sns.swarmplot(x='day',y = 'total_bill', data = dataset)
```

```python
# If we observe above, swarmplot looks good on violin plot, let us do that.
sns.violinplot(x = 'day', y = 'total_bill', data = dataset)
sns.swarmplot(x = 'day', y = 'total_bill', data = dataset, color = 'black')
```

###### Relational plots -  are used for visualizing the statistical relationship between the data points. 

    * Visualization is necessary because it allows the human to see trends and patterns in the data. 
    * The process of understanding how the variables in the dataset relate each other and their relationships are termed as Statistical analysis. 
    * relational plot seaborn provides three functions ::
        
        * relplot()
        * scatterplot()
        * lineplot()


#### relplot() - seaborn.relplot(x=None, y=None, data=None, **kwargs) 

    x,y - numeric / input data variables
    hue, size, style - dataset features(columns name) / Groupig variables that will produce elements with different colors.
    kind - default scatter, or line
    palette - different colors, can be dictionaries, list
    size - plot size


```python
#Visualizing the most basic plot to show all the data points in tips dataset.

sns.set(style ="ticks")  # a neat x, y grid
sns.relplot(x = 'total_bill', y = 'tip', data = dataset)
plt.show()
```

```python
#Grouping data points on the basis of category, here as time.

sns.set(style ="ticks") 
sns.relplot(x = 'total_bill', y = 'tip', hue = 'time' ,data = dataset) 
# plots but with two values Dinner, lunch based on time
plt.show()
```

```python
#using time and sex for determining the facet of the grid.

sns.relplot(x="total_bill", y="tip", hue="day", col="time", row="sex", data=dataset) 
#cols, rows plot two new graph on total_bill vs tips on dinner and lunch each based on categorical values sex
plt.show()
```

```python
# using size attribute, we can see data points having different size

sns.relplot(x="total_bill", y="tip", hue="day", size = "size", data = dataset)
plt.show()

#each day is color coded and the sizes are imdexed and you can see that majority of total amount an tip is concentrated
# between 10 - 20 and 2 - 4 respectively.
```

```python
#### relplot with scatter as kind

sns.relplot(x = 'total_bill', y = 'tip', kind = 'scatter', data = dataset)
plt.show()
```

```python
#### relplot with kind as line

sns.relplot(x = 'total_bill', y = 'tip', kind = 'line', data = dataset)
plt.show()
```
