### Frequency distribution

    * Summary form of data presented in class interval and frequencies.
    * In simple words, frequency distribution is a tabular summary, grouping of the frequencies of the observations in each of several non-overlapping classes. 
    
    * Range: The difference between highest and lowest element in data. 
                Range = highest element value - lowest element value. (If we use pandas.describe() on a dataframe, range is 
                    difference between max and min value. Try it out).
     
    * Class interval should be no less than 5 and no greater than 15. The statistics researcher will decide on the class interval.
    
    * Class midpoint - The value halfway between the lower and the upper case limits. It can be calculated as the average of class end points.
    * Relative frequency - Represented in the form of the proportions and percentages. It is the proportions of the total frequencie for any given classes end points.
    * Cumilative Frequency distribution - It is the proportions of observations within values less than or equal to the upper limit of any class interval.
            - Cumilative frequency for each class interval is the frequency fr that class interval added to the preceding cumilative total.

### Measures of Dispersion

    * Measure of Dispersion - The study is focused on statistical techniques to measure deviation of data value from the 
        central tendency whcih will usually be median or mean.
        
    * Measure of shapes - The study is focused to statistical technique to describe the shape of the distribution.
    
    
* Measure of Dispersion:: 

        * Absolute measure of Dispersion - Absolute measure of dispersion are presented in the same unit as the unit of distribution. If the units are meters of cloth, then the measure of the dispersion will be presented as meters. It is helpful in measuring if the unit are of same type.
        
        * Relative measure of Dispersion - Relative measure of dispersion are most useful in comparing two sets of data which have two different units of measurement. It is expressed as percentage or the co-efficient of the absolute measure of dispersion.
        
        
#### Properties of a good measue of Dispersion:

        A. It should be rigidly defined and simple to understand.
        B. It should be based on all the observations.
        C. Fluctuations of samplings should not affect it.
        D. It should be suitable for further mathematical treatment.
        E. It should be not be affected by extreme values.


#### Methods of measuring dispersion:

        A. Range
        B. Interquartile range and quartile deviation
        C. Mean deviation or average deviation
        D. Standard Deviation, variance and co-efficient of variance.

A. Range

    * Simple measure of dispersion.
    * A difference between biggest and smallest values. 
    * Symbolically R = L - S  ,where L is Largest and S is smallest.
    * The relative measure of dispersion for range is called co-efficient of range and is calculated by ratio of (L-S)/(L+S).
    

B. Interquartile range and quartile deviation

    * Interquartile range is the difference between the third quartile and the first quartile. 
    * Refer above images for the formula for the 1st quartile and 3rd quartile.
    
    * Quartile deviation or semi-interquartile range = (Q3 - Q1) / 2,
                where
                        Q3 and Q1 are the 3rd Quartile and 1st Quartile respectively.
    * Co-efficient of Quartile deviation can be used to measure degree of variation in two different distribution when both
    have different units of measurement.
    
    * Co-efficient of the quartile deviation = (Q3 - Q1) / (Q3 + Q1), where Q3 and Q1 has its meanings.

1. Experiment - A process which produces outcomes. Eg: A coin toss, Tossing a coin is an experiment and head/tail are the outcomes.

2. Event - An outcome of the experiment. Getting a head or tail outcome based on a coin toss experiment is an event.

3. Compound event - A joint occurances of two or more simple events is known as compount event. Eg, two coin which are tossed, ending with an outcome one head and one tail for the coin toss event i compound event

4. Independent and Dependent events 

        * An event is said to be independent if the occurance or non-occurance of one is not affected by occurance or non-occurance of another. Eg: Two seperate coin toss which doesn't affect outcomes of both.
        
        * Two or more events are said to be dependent if occurance of one event influence the occurannce of another. Eg: A pack of cards, if an King of hearts is removed and not replaced, the second event is decided on another card being removed and not replaced.
        
5. Mutually Exclusive events - Two or more events are said to be mutually exclusive if the occurance of one implies that the other cannot occur. Eg: A coin toss, either head or tail outcomes is possible, but never both.

8. Conditional probability - A conditional probability of two events E1 and E2 is generally denoted by P(E1/E2) and is the probability of the occurance of E1 that E2 has already occured. Eg: A probability of person having cold coffee which is E1 given that his coffee already has ice crushed in it E2.

11. Poisson Probability Distribution - a Poisson distribution is a statistical distribution that shows how many times an event is likely to occur within a specified period of time. It is used for independent events which occur at a constant rate within a given interval of time.

Poisson Distribution has 5 requirements :

        * Each occurance of the event is independent of the occurance of the other event.
        * The probabilty of occurance is the same for any two intervals of equal lenght.
        * Poisson distribution describes discrete occurances over a specific time interval.
        * The expected number of occurances must hold constant for all time intervals of the same size.
        * In each interval, occurances ranges from 0 to infinity.
