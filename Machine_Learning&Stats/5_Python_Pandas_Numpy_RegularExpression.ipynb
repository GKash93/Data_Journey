### Pandas DataStructure

    1. series
    2. dataframe
    

### pandas - series

##### An 1-Dimensional array with homogenous data. It should be same datatype.
##### creating a pandas series.

##### dataset_series = pd.Series(data, index, dtype, copy)

    1. data: data takes various forms like ndarray, list, constants
    2. index: Index values must be unique and hashable, same length as data. Default np.arrange(n) if no
    3. index is passed.
    4. dtype: dtype is for data type.
    5. copy: Copy data. Default False


```python
import pandas as pd

data = np.array(['a','b','c','d'])
s = pd.Series(data, index = [100,101,102,103])
s
```

### pandas - DataFrame

##### An 2-Dimensional array with homogenous data. It should be same datatype.
##### creating a pandas dataframe.

##### dataset_dataframe = pd.DataFrame(data, index, columns, dtype, copy)

    1. data: data takes various forms like ndarray, series, map, lists, dict, constants and also another DataFrame.
    2. index: For the row labels, the Index to be used for the resulting frame is Optional Default np.arange(n) if no index            is passed
    3. columns:  For column labels, the optional default syntax is - np.arange(n). This is only true if no index is passed.
    4. dtype: Data type of each column.
    5. copy: This command (or whatever it is) is used for copying of data, if the default is False.

#### Method and attributes of any dataframe

    1. shape - rows and columns of your working dataframe
    2. head() - shows top 5 rows and all the columns in the dataframe. 
        2A. head(10) - shows the top 10 rows and all the columns in the dataframe
    3. tail() - show the last 5 rows and all the columns in the dataframe
        3A. tail(25) - shows the last or bottom 25 rows and all columns in the dataframe.
    4. dtypes - shows all the datatype of the column in the dataframe.
    5. info() - the most important method, which will help us to find how many values we have and how many null values the            dataframe have.

### Indexing, Selecting and Assigning

#### Indexing - Selecting particular rows and columns in the dataframe.
    
    1. loc - selecting data based on the labels
    2. iloc - selecting data based on the numerical position.
    

```python
#iloc

print(dataset.iloc[0])

print(dataset.head(5))
```

```python
print(dataset.iloc[0,0])
```

```python
print(dataset.iloc[0:5, 0: 8]) #rows 0th position to 5th position and columns 0th position to 8th position
```

```python
print(dataset.loc[0,['land_area', 'percent_city']])
```

```python
dataset.land_area #attributes dot based selection
```

```python
dataset[['graduates','work_force']] #dictionary based selection
```

```python
dataset[dataset.graduates >= 70.0] # conditional selection
```

```python
#summary functions

dataset.describe()

#count -total no of values used, #mean, std are statistical concept. #min and max are the values, #25,50,75 are the quartiles
#sometimes, 50% or 2nd quartile is referred as Median.
```

```python
dataset.describe().loc[['count','mean','std','50%']] #Cherry pick only few attributes for easier understanding.
```

```python
dataset[['land_area', 'percent_city']].describe().loc[['count','mean','std','50%']] 
#cherry pick few features and few attributes for easier analysis
```

```python
# aggregation

print(dataset.mean()) #just mean of all the features and their values

print("\n \n",dataset.median())
```

```python
dataset.income.unique() #provides all the unique value of the feature.
```

```python
dataset.crime_rate.value_counts() #This will give you counts of all unique value in the dataframe or dataset feature.
```

```python
#sorting, an important feature to find a feature with highest value to owest value

dataset.sort_values(by = 'crime_rate').head(10) # here the dataframe is displayed by crime rate having lowest to highest
```
